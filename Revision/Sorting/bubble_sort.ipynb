Bubble sort
1. Compare the next element
2. if left element is greater than right element swap them and process goes on
3. bubble sort is an stable algorithm
4. it is an inefficient algorithm as many swap is involved
5. Time Complexity => O(n^2)
6. space complexity => O(1)

```python
def bubble_sort(arr):
    # get the length of the array
    n = len(arr)
    for i in range(n):
        swwapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j+1], arr[j]
                swwapped = True
        if not swwapped:
            break
    return arr

bubble_sort([5,3,2,1,0,-3,-1,8])
```
