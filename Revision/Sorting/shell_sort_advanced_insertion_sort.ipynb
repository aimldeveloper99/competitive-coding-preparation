shell sort

1. sophisticated version of insertion sort
2. shell sort allows for the swapping of element that are far apart
3. The long distance movement helps element get to their correct positions much faster
4. compute the gap of the array by dividing the length by 2
5. compare the element with the gap element
6. check the condition, if true swap them
7. in the final step the gap is 1, which is exactly same as a standard insertion sort, however the list is nearly sorted so it is extremely fast
8. Time complexity => O(n^2)
9. space complexity => O(1)
10. This is unstable algorithm

```python
def shell_sort(arr):
    n = len(arr)
    gap = n // 2
    while gap > 0:
        for i in range(gap, n):
            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:
                arr[j] = arr[j - gap]
                j-= gap
            arr[j] = temp
        gap //= 2 
    return arr
shell_sort([5,3,2,1,0,-3,-1,8])
```
