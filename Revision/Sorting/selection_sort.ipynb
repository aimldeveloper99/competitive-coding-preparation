Selection sort

1. we select the minimum element, swap it to the first position and the process continues
2. selection sort is an unstable algorithm => meaning if two values are of same name/number their position will change
    a. An unstable algorithm is a sorting method that does not guarantee the original relative order of equal elements (keys) is preserved in the sorted output
3. Time complexity = O(n^2)
4. space complexity = O(1)

```python
def selection_sort(arr):
    #first find the length of the array
    n = len(arr)
    for x in range(n):
        #get the position
        position = x
        for y in range(x+1, n):
            if arr[y] < arr[position]:
                #select the lowest element position from the array
                position = y
        #swapping logic, we select the lowest in the array and swap it to the x index
        arr[x], arr[position] = arr[position], arr[x]
    return arr

selection_sort([5,3,2,1,0,-3,-1,8])
```
