

**Contain Duplicates**

Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

Example 1:

Input: nums = [1, 2, 3, 3]

Output: true

Example 2:

Input: nums = [1, 2, 3, 4]

Output: false

```python
from typing import List
def hasDuplicate(nums: List[int]) -> bool:

    #my solution brute force => not efficient time complexity O(n^2) and space complexity O(1)
    # for x in range(len(nums) - 1):
    #     for y in range(x + 1, len(nums)):
    #         if nums[x] == nums[y]:
    #             return True
    # return False

    #using hashset learned from solution => efficient time complexity O(n) and space complexity O(n)
    #i prefer this

    basket = set()
    for x in nums:
        if x in basket:
            return True
        basket.add(x)
    return False

    #comparing with length learned from discussion => efficient time complexity O(n) and space complexity O(n)
    # return len(set(nums)) != len(nums)
```
